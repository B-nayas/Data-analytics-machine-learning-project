```python
##import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.2f}'.format

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score 
from sklearn.metrics import recall_score,cohen_kappa_score, f1_score
import warnings

warnings.filterwarnings('ignore')
import os
import pickle
```

```python
#import xlsx

dataset = pd.read_excel("..//data//clean_data.xlsx", index_col=0)

##check 

dataset
```

# DATA CLEANING

```python
##wrong columns
dataset.rename(columns={'# Bank Accounts Open':'Bank_Accounts_Open', '# Credit Cards Held':'Credit Cards Held','# Homes Owned':'Homes Owned'}, inplace=True)


dataset = dataset.drop(columns = ['Customer Number'])
dataset = dataset.reindex(columns=['Reward','Mailer Type','Income Level','Bank Accounts Open','Overdraft Protection','Credit Rating','Credit Cards Held','Homes Owned','Household Size','Own Your Home','Average Balance','Q1 Balance','Q2 Balance','Q3 Balance','Q4 Balance','Offer Accepted'])

##drop NaN
dataset = dataset.dropna()
##check
dataset
```

```python
dataset.replace({'Letter': 0, 'Postcard': 1}, inplace=True)
```

```python
dataset.replace({'Low': 0, 'Medium': 1, 'High': 2}, inplace=True)
```

# Create scaler 1

```python
X = final_dataset.drop(columns = ['Offer Accepted'])
```

```python
X
```

```python
y = final_dataset['Offer Accepted']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=35)
```

```python
def create_scaler(train_set,name = 'scaler1.pkl'):

    
    if ( os.path.isdir('..//scalers//scaler1.pkl') ):
        scaler = pickle.load(open('scalers/' + name, 'rb'))
    else:
        if ( os.path.isdir('..//scalers/') == False ):
            os.makedirs('..//scalers/')
        scaler = PowerTransformer()
        scaler.fit(train_set)
        pickle.dump(scaler, open('..//scalers/' + name, 'wb'))
        
    return scaler

```

```python

scaler = create_scaler(X_train)
```

```python
final_dataset
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

##Â Model training

```python
def create_logistic_model(train_set_scaled, y_train, name = 'model1.pkl'):
  
    
    if ( os.path.isdir('..//models/model1.pkl') ):
        model = pickle.load(open('..//models/' + name, 'rb'))
    else:
        if ( os.path.isdir('..//models/') == False ):
            os.makedirs('..//models/')
        model = LogisticRegression()
        model.fit(X_train_scaled, y_train)
        pickle.dump(model, open('..//models/' + name, 'wb'))
        
    return model
```

```python
lr = create_logistic_model(X_train_scaled_df, y_train, name = 'model1.pkl')
```

```python
y_train_pred = lr.predict(X_train_scaled)
y_test_pred = lr.predict(X_test_scaled)
```

## Model evaluation

```python
def evaluate_model(y_train, y_train_pred, y_test, y_test_pred):
    
    
    cm_train = confusion_matrix(y_train, y_train_pred)
    cm_test  = confusion_matrix(y_test,  y_test_pred)
    
    results = pd.DataFrame({'Metric': ['accuracy','precision','recall','kappa','f1 score'],
    'Train': [accuracy_score(y_train, y_train_pred), precision_score(y_train, y_train_pred),
             recall_score(y_train, y_train_pred), cohen_kappa_score(y_train, y_train_pred),
             f1_score(y_train, y_train_pred)],
    'Test': [accuracy_score(y_test, y_test_pred), precision_score(y_test, y_test_pred),
             recall_score(y_test, y_test_pred), cohen_kappa_score(y_test, y_test_pred),
             f1_score(y_test, y_test_pred)]})
    results = results.set_index(["Metric"]).T

    
    display(results)
    
    return cm_train, cm_test, results
```

```python
cm_train, cm_test, results = evaluate_model(y_train, y_train_pred, y_test, y_test_pred)
```
